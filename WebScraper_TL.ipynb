```python
l_pages = []
for i in range (1,197):
    page_number = str(i)
    page_link = 'https://flight-report.com/en/s/?page='+ page_number + '&departure_input=&arrival_input=&auteur=&constructor=0&modeles=0&annee=0&Y=1&W=1&J=1&F=1&D=1'
    l_pages.append(page_link)
print (l_pages)
```

```python
l_link = []
for i in range (len(l_pages)):
    hdr = {'User-Agent': 'Mozilla/5.0'}
    req = Request(l_pages[i],headers=hdr)
    page = urlopen(req)
    soup = BeautifulSoup(page)
    page_soup = BeautifulSoup(page, "html.parser")
    bodies = soup.body
    bodies = soup.find_all("article", {"class":"col-xs-12 col-sm-6 col-md-4 col-lg-3 report-article"})
    for i in range (len(bodies)):
        link = bodies[i].a['href']
        link = "https://flight-report.com"+link
        l_link.append(link)
print(l_link)
```

```python
len(l_link)
```

```python
l_full_review = []
for i in range (len(l_link)):
    hdr = {'User-Agent': 'Mozilla/5.0'}
    req = Request(l_link[i],headers=hdr)
    page = urlopen(req)
    soup = BeautifulSoup(page)
    a = soup.find('div', {'id':'data-area'})
    b = a.find_all('span',{'class':"truncate"})
    d = a.find_all('span',{'class':"city-flight"})
    l_key_features = []
    l_grades = []
    l_full_rev = []
    l_conclusion = []
    l_flights = []
    for c in b:
        key_features = c.text.strip()
        l_key_features.append(key_features)
    for e in d:
        dep_arr = e.text.strip()
        l_key_features.append(dep_arr)
    raw_ranking = soup.find_all('div', {'class':'rowRanking'})
    for i in range (len(raw_ranking)):
        global_rank = raw_ranking[i].find_all('span', {'class':'notation'})
        for j in range (len(global_rank)):
            rank = global_rank[j].text
            l_grades.append(rank)
    full_rev = soup.find('div', {'itemprop':'reviewBody'})
    full_rev = full_rev.text.strip('\n\n\n\n\t\t\t\t\t\t(adsbygoogle = window.adsbygoogle || []).push({});\n\t\t\t\t\t\t\n\n\t\t\t\t\t\n\t\t\t\t\t')
    l_full_rev.append(full_rev)
    conclusion_raw = soup.find('section', {'id':'Conclusion'})
    c = conclusion_raw.find_all('p')
    for i in range (len(c)):
        conclusion = c[i].text
        l_conclusion.append(conclusion)
        l_conclusion = list(filter(None, l_conclusion))
    l_conclusion = ','.join(l_conclusion)
    l1_conclusion = [l_conclusion]
    l_flights = l_key_features + l_grades + l_full_rev + l1_conclusion
    l_full_review.append(l_flights)
print (l_full_review)
```
